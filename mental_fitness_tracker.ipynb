```python
df=data.copy()
df.head(10)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in df.columns:
  if df[i].dtype=='object':
    df[i]=l.fit_transform(df[i])
```

```python
x=df.drop('mental_fitness',axis=1)
y=df['mental_fitness']
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.20,random_state=2)
```

```python
print("xtrain: ",xtrain.shape)
print("xtest: ",xtest.shape)
print("\n ytrain: ",ytrain.shape)
print("ytest: ",ytest.shape)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
lr=LinearRegression()
lr.fit(xtrain,ytrain)
ytrain_pred=lr.predict(xtrain)

mse=mean_squared_error(ytrain,ytrain_pred)
rmse=(np.sqrt(mean_squared_error(ytrain,ytrain_pred)))
r2=r2_score(ytrain,ytrain_pred)
print("linear regression model performance for training set:")
print("---------------------------------------------------------------------")
print('mse is {}'.format(mse))
print('rmse is {}'.format(rmse))
print('r2 is {}'.format(r2))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(xtrain,ytrain)
ytrain_pred=rf.predict(xtrain)
print(ytrain_pred)
mse=mean_squared_error(ytrain,ytrain_pred)
rmse=(np.sqrt(mean_squared_error(ytrain,ytrain_pred)))
r2=r2_score(ytrain,ytrain_pred)
print('mse is {}'.format(mse))
print('rmse is {}'.format(rmse))
print('r2 is {}'.format(r2))
```

```python
import numpy as np
import matplotlib.pyplot as plt
lr=LinearRegression()
lr.fit(xtrain,ytrain)
ytrain_pred=lr.predict(xtrain)
time_of_day = np.linspace(0, 20, 5472)
# Calculate the residuals (difference between actual and predicted data)
residuals = ytrain - ytrain_pred

# Plot the residuals
plt.figure(figsize=(10, 6))
plt.plot(time_of_day, residuals, marker='o', linestyle='-', color='b')
plt.axhline(y=0, color='r', linestyle='--', label='Zero Line')  # Zero line for reference
plt.xlabel('Time of Day')
plt.ylabel('Residuals (Difference)')
plt.title('Residual Plot (Difference between Actual and Predicted Mood)')
plt.legend()
plt.grid(True)
plt.show()
```
